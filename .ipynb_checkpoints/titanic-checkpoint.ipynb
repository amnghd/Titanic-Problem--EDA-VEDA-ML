```python
for dataset in combine:
    dataset['has_age']=(~dataset.Age.isnull()).map(has_age_map)
    dataset.Age.fillna(value=combine[0].Age.mean(), inplace=True)
```

**4.1.1- Title (Engineered Feature#1)**

We might be able to extract a feature from the name column (title). Title might show the social and economical level of the sample and might proide some information.We use regular expression to extract this feature.

```python
pattern='([A-Za-z]+)\.' #patern defining only alphabetical values followed by a period "."
for dataframe in combine: #performs it on both groups
    dataframe['Title']=dataframe.Name.str.extract(pattern , expand=False)
train_df.Title.unique()
```

```python
for dataframe in combine:
    dataframe.Title.replace(['Lady','Countess','Capt','Col', 'Don','Dr','Major','Rev','Sir', 'Jonkheer', 'Dona'], 'Rare', inplace=True)
    dataframe.Title.replace(['Mlle','Ms'],'Miss', inplace=True)
    dataframe.Title.replace('Mme','Mrs',inplace=True)
```

```python
train_df[["Title", "Survived"]].groupby(['Title'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
print(train_df.Title.value_counts())
```

Most of the sklearn models work with numerical categories (though they have techniques to make categorical string into numerical values). However, it is a good practice to make them integer in order to make sure data is model fitting ready.

```python
title_map={'Mrs':1, 'Miss':2, 'Master':3, 'Rare':4, 'Mr':5}
for dataframe in combine:
    dataframe['Title']=dataframe.Title.map(title_map,na_action='ignore')
combine[0].Title.unique()
```

Let's quickly see the correlation of the features we have. Before it let's drop the features we extract features from (Name, Cabin, and then map the categorical values into numerical integers to make the learning easier.

```python
sex_map={'female':1,'male':0}
embarked_map={'C':0,'S':1,'Q':2}


combine[0].drop(['PassengerId','PassengerId_Category','Name', 'Ticket', 'Cabin', 'Fare_Category', 'Age_Category'], axis=1, inplace=True)
combine[1].drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
for dataset in combine:
    dataset['Sex']=dataset.Sex.map(sex_map)
    dataset['Embarked']=dataset.Embarked.map(embarked_map)
    dataset['Embarked']=dataset.Embarked.fillna(dataset.Embarked.median())
    
```

```python
Fare_train, bins_train= pd.qcut(train_df.Fare,4,labels=False, retbins=True)
bins_train[0]=0
bins_train[-1]=600
Fare_test=pd.cut(test_df.Fare, bins=bins_train, labels=False, include_lowest=True)
train_df['Fare']=Fare_train
test_df['Fare']=Fare_test
```

### 5.1. Support Vector Machine

We are starting with the support vector machine. SVC requires parameter tunning, since it has a parameter to tune to assuer highest accuracy. Parameter tunning, itself, is a modelling step. Therefore, we cannot get the model accuracy using test set. We need to have the train set split into a train and cross validation set, then we choose on which parameter our model does best.
For this purpose, we are using GridSearchCV from sklearn.
This deals with the C $(1/\lambda)$ and kernel (familiarity function).

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
parameters={'kernel':['linear','rbf'],'C':[0.1,.3,1,3,10,30]}
svc=SVC()
clf_svc=GridSearchCV(estimator=svc,param_grid=parameters, cv=5)
clf_svc.fit(X_train, Y_train)
print("Best parameters are {}".format(clf_svc.best_params_))
print("Best score is {}".format(clf_svc.best_score_))
```

```python
acc_svc=clf_svc.best_score_*100
acc_svc
```

Before going a step further. let's see if reducing the dimentionality might improve our model. Maybe there are som redunducnay that over confidents the model.
We need to import principle componant analysis, an unsuervised machine learning technique.
When we add PCA as a parameter, we need to put SVC and PCA into a pipeline.

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
pl=Pipeline([('pca',PCA()), ('svc',SVC())])
parameters={'pca__n_components':[1,3,6,9,11], 'svc__kernel':['linear','rbf'],'svc__C':[0.1,.3,1,3,10,30]}
clf=GridSearchCV(estimator=pl,param_grid=parameters, cv=5)
clf.fit(X_train, Y_train)
print("Best parameters are {}".format(clf.best_params_))
print("Best score is {}".format(clf.best_score_))
```

```python
acc_reduced=100*clf.best_score_
acc_reduced
```

It is obvious, no improvement from dimentionality reduction. Now lets normalize the features. The model prefers 11 features (the whole dimentionality).

```python
from sklearn.preprocessing import StandardScaler
pl=Pipeline([('scl',StandardScaler()), ('svc',SVC())])
parameters={'svc__kernel':['linear','rbf'],'svc__C':[0.1,.3,1,3,10,30]}
clf=GridSearchCV(estimator=pl,param_grid=parameters, cv=5)
clf.fit(X_train, Y_train)
print("Best parameters are {}".format(clf.best_params_))
print("Best score is {}".format(clf.best_score_))
```

```python
acc_scaled=clf.best_score_*100
```

No improvements here, too. It seems odd, scaler is supposed to improve the efficiency of SVC as the similarity function compares the eulicidian distance. Although similar effect is refered [here](https://stackoverflow.com/questions/26171335/why-does-my-svms-performance-drop-after-scaling-the-training-and-test-data), using scaler is suggested for SMV.

### 5.5. Random Forest (Classifier, and also feature importance selection)
We use random forest for classification. Moreover, it can help us a power to take a look at the features and their powers.

```python
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint as sp_randint

rf=RandomForestClassifier()
param_grid = {"max_depth": [3, None],
              "max_features": [1, 3, 10],
              "min_samples_split": [2, 3, 10],
              "min_samples_leaf": [1, 3, 10],
              "bootstrap": [True, False],
              "criterion": ["gini", "entropy"]}

clf_rf = GridSearchCV(rf, param_grid=param_grid)
clf_rf.fit(X_train, Y_train)
print("Best parameters are {}".format(clf_rf.best_params_))
print("Best score is {}".format(clf_rf.best_score_))

```

```python
acc_rfo=clf_rf.best_score_*100
acc_rfo
```

Alright, we hae imporoved the method $0.4\%$ from $83.2$ to $83.6$ . Let's see how much each feature is strong in increasing the purity of classes (information gain). One way is to use fature importance attribute of random forest as follows.
Please note that since each classifier find the purity in its own specific way, not necessarily all the classifiers result in similar feature importance. Moreover, since many classifier use non-linear feature, it is possible that a combination, or a power of these faetures show more powerful variance with the target. This represnetation just gives us a basic understanding of whether our feature engineering is working properly.

```python
forest=RandomForestClassifier(n_estimators=10000, random_state=0, n_jobs=-1)
forest.fit(X_train, Y_train)
feat_labels=X_train.columns
importance=forest.feature_importances_
indices=np.argsort(importance)[::-1]
for f in range(X_train.shape[1]):
    print ("{}){}    {}".format(f+1,feat_labels[indices[f]], importance[indices[f]]))
```
